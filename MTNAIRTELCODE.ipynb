```python
#This program tries to show how loading credit on phones(airtime) works especially in Africa where Airtel and 
#and MTN operates

def mtn_airtel_program():
    
    mtn_airtel = input('\nDo you want to load airtime on mtn (m) or airtel (a)?, \n\nKindly type either (m) or (a) and then press the Enter key: ' )
    if mtn_airtel == 'm':
        
        code = int(input('\nPlease enter the code *160# or another number for other options: '))
        if code == 150 :
            print('\nPlease you have entered a wrong code, this is for mobile data not airtime')
            
        elif code == 160 :
            print('\nPlease you have entered a correct code, continue')
            mtn_options = ['Buy airtime','Send airtime','Send Me2U']
            print('\nSelect an option of your choice\n')
            for i in range (len(mtn_options)):
                print(i,'.', mtn_options[i])
            
            mtn_option = input('\nEnter the number of your option: ')
            if mtn_option == '0' :
                print('\nYou are going to buy airtime, and the card should be 15 digits')
                airtime = input('\nEnter the airtime card number: ')
                if len(airtime)==15:
                    print('\nYou have loaded your airtime correctly')
                else:
                    print('\nYour card is incorrect and airtime is not loaded')
            
            if mtn_option == '1' :
                print('\nYou are going to Send airtime, Please enter the 10 digit phone number below and press Enter button')
                Phone_number = input('\nEnter the the phone number to send airtime to and enter: ')
                
                if len(Phone_number)==10:
                    print('\nYou have entered a correct phone number your')
                else:
                    print('\nYour phone number is incorrect, please try again')
            
            if mtn_option == '2' :
                print('\nYou are going to Send Send Me2U, Please enter the 10 digit phone number below and press Enter button')
                Phone_number = input('\nEnter the the phone number to Send Me2U to and enter: ')
                
                if len(Phone_number)==10:
                    print('\nYou have entered a correct phone number')
                    Me2U_airtime = input('\nEnter the airtime amount: ')
                    
                    if len(Me2U_airtime)>0:
                        print('\nYou have entered a correct amount')
                    else:
                        print('\nYour airtime amount loaded is incorrect')
                else:
                    print('\nYour phone number is incorrect')
        
        else:
            print('\nPlease try again later')  
    
    
    ########################################################################################################################
    if mtn_airtel == 'a':
        
        code = int(input('\nPlease enter the code *151# or another number for other options: '))
        if code == 150 :
            print('\nPlease you have entered a wrong code, this is for mobile data not airtime')
            
        elif code == 151 :
            print('\nPlease you have entered a correct code, continue')
            airtel_options_list = ['Buy airtime','Send airtime','Send Me2U']
            print('\nSelect an option of your choice\n')
            for i in range (len(airtel_options_list)):
                print(i,'.', airtel_options_list[i])
            
            airtel_options = input('\nEnter the number of your option: ')
            
            if airtel_options == '0' :
                print('\nYou are going to buy airtel airtime')
                airtime = input('\nEnter the airtime card number, which is 14 digits: ')
                if len(airtime)==14:
                    print('\nYou have loaded your airtime correctly')
                else:
                    print('\nYour card is incorrect and airtime not loaded')
            
            if airtel_options == '1' :
                print('\nYou are going to Send airtime, Please enter the 10 digit phone number below and press Enter button')
                Phone_number = input('\nEnter the the phone number to send airtime to and enter: ')
                
                if len(Phone_number)==10:
                    print('\nYou have entered a correct phone number your')
                else:
                    print('\nYour phone number is incorrect')
            
            if airtel_options == '2' :
                print('\nYou are going to Send Send Me2U, Please enter the 10 digit phone number below and press Enter button')
                Phone_number = input('\nEnter the the phone number to Send Me2U to and enter: ')
                
                if len(Phone_number)==10:
                    print('\nYou have entered a correct phone number!')
                    Me2U_airtime = int(input('\nEnter the airtime amount: '))
                    
                    if Me2U_airtime>0:
                        print('\nYou have entered a correct amount of airtime')
                    else:
                        print('\nYour airtime amount loaded is incorrect')
                else:
                    print('\nYour phone number is incorrect')
        
        else:
            print('\nPlease try again later') 
    
    else:
        print('\nYou entered an unknown choice,', mtn_airtel, '\n\nThe correct input should be either m or a')
    
    
mtn_airtel_program()
```
