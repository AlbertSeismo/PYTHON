```python
from scipy import *
from matplotlib.pyplot import *
from numpy import *
#from sklearn import *
from seaborn import *
set()
```

```python
%matplotlib notebook

figure(figsize=(9,5))
x = arange(-5,5,0.5)
y = 2*x+5

scatter(x,y)
title('A plot of y against x \n')
xlabel('x-data')
ylabel('y-data')
```

```python
%matplotlib notebook

figure(figsize=(9,5))
x = arange(-5,5,0.5)
y = 2*x**2+5*x-1

scatter(x,y)
title('A plot of y against x \n')
xlabel('x-data')
ylabel('y-data')
legend(['y = 2*x**2+5*x-1'],loc ='best')
```

```python
%matplotlib notebook

figure(figsize=(9,5))
x = arange(-5,5,0.5)
y = 3*x**3+5*x-10

plot(x,y)
title('A plot of y against x \n')
xlabel('x-data')
ylabel('y-data')
legend(['y = 3*x**3+5*x-10'],loc ='best')
```

```python
%matplotlib notebook

figure(figsize=(9,5))

x = arange(0,30,0.1)
y1 = cos(x)
y2 = sin(x)
plot(x,y1)
stem(x,y2)
legend([['y1'],['y2']], loc='best')
title('A plot of sine and cosine against the angle (thetha)\n')
xlabel('Angle(Degrees)')
ylabel('Y')
```

```python
%matplotlib notebook

figure(figsize=(9,5))

x = arange(0,60,0.005)
y1 = 0.5*cos(x/2)
y2 = 3*sin(3*x+1)
Y = y1+y2
plot(x,Y)
legend([['Y']], loc='best')
title('A plot of trig combinations against the angle (thetha)\n')
xlabel('Angle(Degrees)')
ylabel('Y')
```

```python
%matplotlib notebook

figure(figsize=(9,5))
x = arange(0,5,0.005)
y = exp(-x)*sin(10*x-1)

plot(x,y)
legend(['y= exp(-x)*sin(10*x-1)'], loc='best')
title('A plot of a damped oscillator against the angle (thetha)\n')
xlabel('Angle(Degrees)')
ylabel('Y')
```

```python
%matplotlib notebook

figure(figsize=(9,5))

x = arange(0,10,0.1)
y = tan(x)
plot(x,y)
legend(['y=tan(x)'], loc='best')
title('A plot of tangent of an angle against the angle (thetha)\n')
xlabel('Angle(Degrees)')
ylabel('Y')
```

```python
%matplotlib notebook
#random.seed(0)
x = arange(0,5,0.05)
y = 2*exp(-(x-3)**2/0.4)

figure(figsize=(8,5))
subplot(1,3,1)
plot(x,y)
xlabel('x-data')
ylabel('y-data')
title('A graph of y against x\n')

subplot(1,3,3)
scatter(x,y)
xlabel('x-data')
ylabel('y-data')
title('A graph of y against x\n')
```

```python
%matplotlib notebook
#random.seed(0)
x = arange(0.5,5,0.5)
y = log((x-0.1)/(x+1))

figure(figsize=(8,5))
subplot(3,1,1)
stem(x,y)
xlabel('x-data')
ylabel('y-data')
title('A graph of y against x\n')

subplot(3,1,2)
scatter(x,y)
xlabel('x-data')
ylabel('y-data')
title('A graph of y against x\n')

subplot(3,1,3)
plot(x,y)
xlabel('x-data')
ylabel('y-data')
title('A graph of y against x\n')
```
